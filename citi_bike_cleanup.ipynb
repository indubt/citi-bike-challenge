```python
# Import dependencies
import pandas as pd
import zipfile
import os
import glob 
from pathlib import Path
```

```python
# Open the zip file
with zipfile.ZipFile('resources/202401-citibike-tripdata.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('resources/unzip')

with zipfile.ZipFile('resources/202402-citibike-tripdata.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('resources/unzip')

with zipfile.ZipFile('resources/202403-citibike-tripdata.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('resources/unzip')
```

```python
# Get all the csv files
csv_files = os.path.join("resources/unzip", "*.csv") 
  
# glob is used to get all files that end with .csv 
all_files = glob.glob(csv_files) 
  
# read from csv files 
citibike_df = pd.concat(map(pd.read_csv, all_files), ignore_index=True) 
citibike_df.head()
```

```python
print(f'length of combined csv: ${len(citibike_df)}')
```

```python
citibike_df.to_csv('resources/output/citibike.csv')

# Export to CSV 
filepath = Path('resources/output/citibike.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
citibike_df.to_csv(filepath, index=False)  
```
