```python
# Import dependencies
import pandas as pd
import zipfile
import os
import glob 
from pathlib import Path
import random
import datetime as dt
```

```python
# path that contains the rawdata zip files downloaded from https://citibikenyc.com/system-data
zipfiles_Path = Path("resources/rawdata") 

# for each zipfile in the given path
for file in zipfiles_Path.glob('*.zip'):

    # extract the contents of the zipfile in a directory
    with zipfile.ZipFile(file, 'r') as zip_ref:
        zip_ref.extractall('resources/unzip')
```

```python
# path that contains the csv files
unzip_path = "resources/unzip/"
extension = "*.csv"

# get the path for all the csv files 
all_csv_files = [file 
                 for path, subdir, files in os.walk(unzip_path)
                 for file in glob.glob(os.path.join(path, "*.csv"))]
                 
print(all_csv_files)
```

```python
# Create a function to pick sample data from each csv file's dataframe
def random_sample(df):

    # select a random list of 2000 indices from the length of the dataframe
    random_list = random.sample(range(0, len(df)), 2000)

    # create the sample dataframe with 2000 records from the original csv dataframe
    sample_df = df[df.index.isin(random_list)]

    # reset the index
    sample_df = sample_df.reset_index(drop = True)

    # return the sample dataframe
    return sample_df
```

```python
# function to read data from csv , take random sample and return the df
def read_csv_data(file_path):

    # read from csv file
    df = pd.read_csv(file_path)

    # Rename columns
    df = df.rename(columns= {"start station id":"start_station_id", 
                             "start station name": "start_station_name", 
                             "start station latitude": "start_lat", 
                             "start station longitude": "start_lng",
                             "end station id":"end_station_id", 
                             "end station name": "end_station_name", 
                             "end station latitude": "end_lat", 
                             "end station longitude": "end_lng",
                             "bikeid": "ride_id",
                             "usertype": "member_casual",
                             "starttime":"started_at",
                             "stoptime":"ended_at"
                             })

    # select random 2000 samples
    sample_df = random_sample(df)

    # return the df
    return sample_df
```

```python
# create dataframe that holds the sample data from all the csv files
citi_bike_sample_data_df = pd.DataFrame()
```

```python
# for each file in the list of csv files
for csvfile in all_csv_files:
    by_month_df = read_csv_data(csvfile)
    citi_bike_sample_data_df = pd.concat([citi_bike_sample_data_df, by_month_df])
    citi_bike_sample_data_df = citi_bike_sample_data_df.reset_index(drop = True)
```

```python
citi_bike_sample_data_df.head()
```

```python
# drop unwanted columns
citi_bike_clean_data_df = citi_bike_sample_data_df.drop(columns= {"birth year", "gender", "start_station_id", "end_station_id", "ride_id", "tripduration"})

# update data in member_casual / usertype to match latest dataset
citi_bike_clean_data_df["member_casual"] = citi_bike_clean_data_df["member_casual"].map({"Subscriber": "member", "Customer": "casual", "member": "member", "casual": "casual"})

# update rideable_type to unknown for older dataset:
citi_bike_clean_data_df["rideable_type"] = citi_bike_clean_data_df["rideable_type"].map({"NaN": "Unknown", 
"docked_bike": "docked_bike", 
"classic_bike": "classic_bike",
"electric_bike": "electric_bike"
})

# round latitude and longitude to 4 digits
citi_bike_clean_data_df["start_lat"] = citi_bike_clean_data_df["start_lat"].round(4)
citi_bike_clean_data_df["start_lng"] = citi_bike_clean_data_df["start_lng"].round(4)
citi_bike_clean_data_df["end_lat"] = citi_bike_clean_data_df["end_lat"].round(4)
citi_bike_clean_data_df["end_lng"] = citi_bike_clean_data_df["end_lng"].round(4)

```
